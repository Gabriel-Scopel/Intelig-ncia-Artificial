```python
import math
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Importa e converte a imagem para RGB
img = cv2.imread('Aviao.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convertendo para preto e branco (RGB -> Gray Scale -> BW)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
a = img_gray.max()
_, thresh = cv2.threshold(img_gray, a/2*1.7, a, cv2.THRESH_BINARY_INV)

# Aplicando operação morfológica de abertura
tamanhoKernel = 5
kernel = np.ones((tamanhoKernel,tamanhoKernel), np.uint8)
thresh_open = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# Filtro de ruído (blurring)
img_blur = cv2.blur(img_gray, ksize=(tamanhoKernel,tamanhoKernel))

# Detecção de borda com Canny (sem blurry)
edges_gray = cv2.Canny(image=img_gray, threshold1=a/2, threshold2=a/2)

# Detecção de borda com Canny (com blurry)
edges_blur = cv2.Canny(image=img_blur, threshold1=a/2, threshold2=a/2)

# Encontrando contornos
contours, hierarchy = cv2.findContours(
    image = thresh,
    mode = cv2.RETR_TREE,
    method = cv2.CHAIN_APPROX_SIMPLE
)
contours = sorted(contours, key = cv2.contourArea, reverse = True)
img_copy = img.copy()
final = cv2.drawContours(img_copy, contours, contourIdx = -1,
                         color = (255, 0, 0), thickness = 2)

# Preparando para plotar as imagens
imagens = [img, img_blur, img_gray, edges_gray, edges_blur, thresh, thresh_open, final]
formatoX = math.ceil(len(imagens)**.5)
if (formatoX**2 - len(imagens)) > formatoX:
    formatoY = formatoX - 1
else:
    formatoY = formatoX

# Loop para plotar todas as imagens, exceto a última
for i in range(len(imagens) - 1):
    plt.subplot(formatoY, formatoX, i + 1)
    plt.imshow(imagens[i], 'gray')
    plt.xticks([]), plt.yticks([])

# Ampliar a última imagem
plt.figure(figsize=(8, 8))  # Define o tamanho da figura para a última imagem
plt.imshow(imagens[-1], 'gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

### Explicação do código:
O código acima realiza uma série de operações de processamento de imagem usando bibliotecas como OpenCV, Numpy e Matplotlib. Inicialmente, a imagem é lida e convertida de BGR para RGB. Em seguida, é transformada para escala de cinza, e uma binarização é aplicada para separar os pixels em preto e branco. Uma operação morfológica de abertura é realizada para remover ruídos, seguida de um filtro de desfoque (blurring) que suaviza a imagem. A detecção de bordas é feita através do algoritmo de Canny, tanto na imagem original quanto na desfocada. Contornos são identificados e desenhados sobre a imagem original, destacando os objetos presentes. Finalmente, todas as imagens processadas são exibidas lado a lado, com a imagem final contendo os contornos ampliada para maior clareza. O código mostra um pipeline básico de técnicas de pré-processamento, detecção de bordas e contornos, comuns em aplicações de visão computacional.
