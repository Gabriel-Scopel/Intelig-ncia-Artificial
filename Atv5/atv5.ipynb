```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

arquivo = np.load('teste2.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])

def run_simulation(hidden_layers, n_runs=10):
    errors = []
    best_model = None
    best_loss_curve = None
    best_y_est = None

    for _ in range(n_runs):
        regr = MLPRegressor(hidden_layer_sizes=hidden_layers,
                            max_iter=1000,
                            activation='relu',
                            solver='adam',
                            learning_rate='adaptive',
                            n_iter_no_change=50)

        regr.fit(x, y)

        y_est = regr.predict(x)

        mse = mean_squared_error(y, y_est)
        errors.append(mse)

        if best_model is None or mse < min(errors):
            best_model = regr
            best_loss_curve = regr.loss_curve_
            best_y_est = y_est
    mean_error = np.mean(errors)
    std_error = np.std(errors)
    
    return mean_error, std_error, best_model, best_loss_curve, best_y_est

# Simulação 1: Arquitetura Simples (1 camada oculta com 5 neurônios)
mean_error_simple, std_error_simple, model_simple, loss_simple, y_est_simple = run_simulation(hidden_layers=(5))

# Simulação 2: Arquitetura Média (2 camadas ocultas com 10 neurônios)
mean_error_medium, std_error_medium, model_medium, loss_medium, y_est_medium = run_simulation(hidden_layers=(10, 10))

# Simulação 3: Arquitetura Complexa (3 camadas ocultas com 20 neurônios)
mean_error_complex, std_error_complex, model_complex, loss_complex, y_est_complex = run_simulation(hidden_layers=(20, 20, 20))

# Escolhe o melhor modelo (aquele com o menor erro médio)
best_mean_error = min(mean_error_simple, mean_error_medium, mean_error_complex)

if best_mean_error == mean_error_simple:
    best_loss = loss_simple
    best_y_est = y_est_simple
    best_title = "Simulação Simples (5 neurônios)"
elif best_mean_error == mean_error_medium:
    best_loss = loss_medium
    best_y_est = y_est_medium
    best_title = "Simulação Média (10, 10 neurônios)"
else:
    best_loss = loss_complex
    best_y_est = y_est_complex
    best_title = "Simulação Complexa (20, 20, 20 neurônios)"

# Apresentando os resultados
print(f"Simulação 1 - Simples (5 neurônios): Média do erro = {mean_error_simple:.4f}, Desvio Padrão = {std_error_simple:.4f}")
print(f"Simulação 2 - Médio (10, 10 neurônios): Média do erro = {mean_error_medium:.4f}, Desvio Padrão = {std_error_medium:.4f}")
print(f"Simulação 3 - Complexo (20, 20, 20 neurônios): Média do erro = {mean_error_complex:.4f}, Desvio Padrão = {std_error_complex:.4f}")
print(f"\nMelhor Modelo: {best_title} com média de erro = {best_mean_error:.4f}")

# Plotando os gráficos para o melhor modelo
plt.figure(figsize=[14,7])

# Gráfico 1: Curva de perda
plt.subplot(1,2,1)
plt.plot(best_loss, label='Perda')
plt.title(f"Curva de Perda - {best_title}")
plt.xlabel("Iterações")
plt.ylabel("Perda (Loss)")
plt.legend()

# Gráfico 2: Comparação entre saída real e estimada
plt.subplot(1,2,2)
plt.plot(x, y, label='Original', color='yellow', linewidth=1)
plt.plot(x, best_y_est, label='Estimado', color='blue', linewidth=2)
plt.title(f"Comparação - {best_title}")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

- Neste experimento, testamos três diferentes arquiteturas de redes neurais artificiais (RNAs) para aproximação de funções, utilizando o MLPRegressor da biblioteca scikit-learn. As arquiteturas variaram em complexidade, desde uma rede simples com uma camada oculta contendo 5 neurônios, até uma rede mais complexa com três camadas ocultas e 20 neurônios em cada. Cada simulação foi executada 10 vezes, e calculamos a média e o desvio padrão do erro final (MSE) para cada arquitetura.

- Principais Observações:
Aquiteturas Simples vs. Complexas:

1) A rede com arquitetura mais simples (uma camada oculta com 5 neurônios) teve desempenho razoável, mas apresentou um erro médio maior em comparação às redes mais complexas. Isso indica que, para o problema em questão, 5 neurônios não são suficientes para capturar toda a complexidade dos dados.

2) A rede mais complexa, com três camadas ocultas de 20 neurônios cada, apresentou o melhor desempenho, com o menor erro médio e o menor desvio padrão entre todas as simulações. Isso sugere que aumentar a capacidade da rede (mais neurônios e mais camadas) foi vantajoso para o ajuste preciso aos dados.

- Estabilidade do Treinamento:

Redes mais simples tiveram um desvio padrão ligeiramente maior, o que significa que suas performances variaram mais entre diferentes execuções. Em contrapartida, a rede mais complexa mostrou uma variação menor (menor desvio padrão), o que indica maior estabilidade no treinamento e menor sensibilidade a variações nos pesos iniciais.
Curva de Perda (Loss Curve):

A curva de perda do melhor modelo (rede complexa) demonstrou uma convergência mais estável e rápida, atingindo um erro de treinamento significativamente menor do que as redes mais simples. Isso indica que essa arquitetura foi mais eficiente em aprender a relação entre as variáveis de entrada e saída.
Comparação entre Valores Reais e Previstos:

O gráfico de comparação entre os valores reais e as previsões do melhor modelo mostrou que a rede conseguiu capturar bem a tendência dos dados. Mesmo com algumas discrepâncias menores, a rede foi capaz de se aproximar dos valores verdadeiros de forma bastante satisfatória.
- Conclusão Final:
Redes neurais com arquiteturas mais complexas (mais camadas e neurônios) tendem a oferecer melhor desempenho em problemas de aproximação de função, desde que o conjunto de dados forneça informação suficiente para evitar o overfitting. Embora as arquiteturas mais simples possam ser adequadas para problemas mais simples ou com dados limitados, as redes mais profundas demonstram maior capacidade de generalização e estabilidade em cenários mais complexos.

No entanto, é importante observar que redes muito grandes podem levar mais tempo de treinamento e demandar maior capacidade computacional. Portanto, o tamanho e a complexidade da rede devem ser balanceados com o volume e a complexidade dos dados disponíveis. O ideal é sempre testar diferentes configurações e ajustar a rede com base nos resultados de erro e na convergência observada.

Em resumo, a rede complexa foi a que apresentou os melhores resultados, sendo uma boa escolha para o problema testado, mas é importante considerar também o custo computacional associado a esse tipo de rede.
