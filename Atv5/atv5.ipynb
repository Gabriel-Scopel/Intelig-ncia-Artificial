```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

arquivo = np.load('teste2.npy')
x = arquivo[0]
y = np.ravel(arquivo[1])

def run_simulation(hidden_layers, n_runs=10):
    errors = []
    best_model = None
    best_loss_curve = None
    best_y_est = None

    for _ in range(n_runs):
        regr = MLPRegressor(hidden_layer_sizes=hidden_layers,
                            max_iter=1000,
                            activation='relu',
                            solver='adam',
                            learning_rate='adaptive',
                            n_iter_no_change=50)

        regr.fit(x, y)

        y_est = regr.predict(x)

        mse = mean_squared_error(y, y_est)
        errors.append(mse)

        if best_model is None or mse < min(errors):
            best_model = regr
            best_loss_curve = regr.loss_curve_
            best_y_est = y_est
    mean_error = np.mean(errors)
    std_error = np.std(errors)
    
    return mean_error, std_error, best_model, best_loss_curve, best_y_est

# Simulao 1: Arquitetura Simples (1 camada oculta com 5 neurnios)
mean_error_simple, std_error_simple, model_simple, loss_simple, y_est_simple = run_simulation(hidden_layers=(5))

# Simulao 2: Arquitetura Mdia (2 camadas ocultas com 10 neurnios)
mean_error_medium, std_error_medium, model_medium, loss_medium, y_est_medium = run_simulation(hidden_layers=(10, 10))

# Simulao 3: Arquitetura Complexa (3 camadas ocultas com 20 neurnios)
mean_error_complex, std_error_complex, model_complex, loss_complex, y_est_complex = run_simulation(hidden_layers=(20, 20, 20))

# Escolhe o melhor modelo (aquele com o menor erro mdio)
best_mean_error = min(mean_error_simple, mean_error_medium, mean_error_complex)

if best_mean_error == mean_error_simple:
    best_loss = loss_simple
    best_y_est = y_est_simple
    best_title = "Simulao Simples (5 neurnios)"
elif best_mean_error == mean_error_medium:
    best_loss = loss_medium
    best_y_est = y_est_medium
    best_title = "Simulao Mdia (10, 10 neurnios)"
else:
    best_loss = loss_complex
    best_y_est = y_est_complex
    best_title = "Simulao Complexa (20, 20, 20 neurnios)"

# Apresentando os resultados
print(f"Simulao 1 - Simples (5 neurnios): Mdia do erro = {mean_error_simple:.4f}, Desvio Padro = {std_error_simple:.4f}")
print(f"Simulao 2 - Mdio (10, 10 neurnios): Mdia do erro = {mean_error_medium:.4f}, Desvio Padro = {std_error_medium:.4f}")
print(f"Simulao 3 - Complexo (20, 20, 20 neurnios): Mdia do erro = {mean_error_complex:.4f}, Desvio Padro = {std_error_complex:.4f}")
print(f"\nMelhor Modelo: {best_title} com mdia de erro = {best_mean_error:.4f}")

# Plotando os grficos para o melhor modelo
plt.figure(figsize=[14,7])

# Grfico 1: Curva de perda
plt.subplot(1,2,1)
plt.plot(best_loss, label='Perda')
plt.title(f"Curva de Perda - {best_title}")
plt.xlabel("Iteraes")
plt.ylabel("Perda (Loss)")
plt.legend()

# Grfico 2: Comparao entre sada real e estimada
plt.subplot(1,2,2)
plt.plot(x, y, label='Original', color='yellow', linewidth=1)
plt.plot(x, best_y_est, label='Estimado', color='blue', linewidth=2)
plt.title(f"Comparao - {best_title}")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

- Neste experimento, testamos trs diferentes arquiteturas de redes neurais artificiais (RNAs) para aproximao de funes, utilizando o MLPRegressor da biblioteca scikit-learn. As arquiteturas variaram em complexidade, desde uma rede simples com uma camada oculta contendo 5 neurnios, at uma rede mais complexa com trs camadas ocultas e 20 neurnios em cada. Cada simulao foi executada 10 vezes, e calculamos a mdia e o desvio padro do erro final (MSE) para cada arquitetura.

- Principais Observaes:
Aquiteturas Simples vs. Complexas:

1) A rede com arquitetura mais simples (uma camada oculta com 5 neurnios) teve desempenho razovel, mas apresentou um erro mdio maior em comparao s redes mais complexas. Isso indica que, para o problema em questo, 5 neurnios no so suficientes para capturar toda a complexidade dos dados.

2) A rede mais complexa, com trs camadas ocultas de 20 neurnios cada, apresentou o melhor desempenho, com o menor erro mdio e o menor desvio padro entre todas as simulaes. Isso sugere que aumentar a capacidade da rede (mais neurnios e mais camadas) foi vantajoso para o ajuste preciso aos dados.

- Estabilidade do Treinamento:

Redes mais simples tiveram um desvio padro ligeiramente maior, o que significa que suas performances variaram mais entre diferentes execues. Em contrapartida, a rede mais complexa mostrou uma variao menor (menor desvio padro), o que indica maior estabilidade no treinamento e menor sensibilidade a variaes nos pesos iniciais.
Curva de Perda (Loss Curve):

A curva de perda do melhor modelo (rede complexa) demonstrou uma convergncia mais estvel e rpida, atingindo um erro de treinamento significativamente menor do que as redes mais simples. Isso indica que essa arquitetura foi mais eficiente em aprender a relao entre as variveis de entrada e sada.
Comparao entre Valores Reais e Previstos:

O grfico de comparao entre os valores reais e as previses do melhor modelo mostrou que a rede conseguiu capturar bem a tendncia dos dados. Mesmo com algumas discrepncias menores, a rede foi capaz de se aproximar dos valores verdadeiros de forma bastante satisfatria.
- Concluso Final:
Redes neurais com arquiteturas mais complexas (mais camadas e neurnios) tendem a oferecer melhor desempenho em problemas de aproximao de funo, desde que o conjunto de dados fornea informao suficiente para evitar o overfitting. Embora as arquiteturas mais simples possam ser adequadas para problemas mais simples ou com dados limitados, as redes mais profundas demonstram maior capacidade de generalizao e estabilidade em cenrios mais complexos.

No entanto,  importante observar que redes muito grandes podem levar mais tempo de treinamento e demandar maior capacidade computacional. Portanto, o tamanho e a complexidade da rede devem ser balanceados com o volume e a complexidade dos dados disponveis. O ideal  sempre testar diferentes configuraes e ajustar a rede com base nos resultados de erro e na convergncia observada.

Em resumo, a rede complexa foi a que apresentou os melhores resultados, sendo uma boa escolha para o problema testado, mas  importante considerar tambm o custo computacional associado a esse tipo de rede.
